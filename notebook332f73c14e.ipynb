```python
from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.layers import Dense
from keras.preprocessing.image import ImageDataGenerator


mod = Sequential()
mod.add(Conv2D(32, (3, 3), input_shape = (50, 50, 3), activation = 'relu'))

mod.add(MaxPooling2D(pool_size = (2, 2)))
mod.add(Conv2D(32, (3, 3), activation = 'relu'))
mod.add(MaxPooling2D(pool_size = (2, 2)))


mod.add(Flatten())
mod.add(Dense(units = 128, activation = 'relu'))
mod.add(Dense(units = 1, activation = 'sigmoid'))


mod.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])


train_datagen = ImageDataGenerator(rescale = 1./255,shear_range = 0.2,zoom_range = 0.2,horizontal_flip = True)
training_set = train_datagen.flow_from_directory('../input/cat-and-dog-train',target_size = (50, 50),batch_size = 32,class_mode = 'binary')
 
mod.fit_generator(training_set,steps_per_epoch = 4000,epochs = 25,validation_steps = 2000)

print("task done")
```
